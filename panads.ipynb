```python
! pip install pandas
import mysql.connector
import pandas as pd
from tabulate import tabulate

# Connect to the MySQL server and the 'library' database
conn = mysql.connector.connect(user='root', password='root', unix_socket='/var/run/mysqld/mysqld.sock', database='library')
cursor = conn.cursor()

# Execute the SELECT query
cursor.execute("SELECT * FROM books")

# Fetch all the results
records = cursor.fetchall()
ids=[]
title=[]
author=[]
year=[]
# Print the records
for record in records:
    ids.append(record[0])
    title.append(record[1])
    author.append(record[2])
    year.append(record[3])

dtyr={
    "id":ids,
    "titles":title,
    "author":author,
    "year":year
}

vr=pd.DataFrame(dtyr)
vr_filtered = vr [(vr['title'] !='')]
new_df = vr.dropna(subset=['title'],inplace=False)
print(new_df)
print(vr_filtered)
print(tabulate(new_df, headers='keys', tablefmt='psql'))
# Close the cursor and connection
cursor.close()
conn.close()
```
